# Data Preparation & Visualization CNN_4rows

## 1. Load data
### 1.1 Load Original and "Faked" Train data

```python
# load first batch of original train data
data_raw_original_1 = pd.read_csv("dataset_2020-02-19/eband_original_collected_train_data_2019-10-20.csv")
print("Shape of data_raw_original_1:", data_raw_original_1.shape)

# load second (current) batch of original train data
data_raw_original_2 = pd.read_csv("dataset_2020-02-19/eband_original_collected_train_data.csv")
print("Shape of data_raw_original_2:", data_raw_original_2.shape)

# append 1 and 2
data_raw_original = data_raw_original_1.append(data_raw_original_2)
print("Shape of data_raw_original:", data_raw_original.shape)

##################################################################

# load augmanted / faked data
data_raw_faked = pd.read_csv("dataset_2020-02-19/eband_faked_train_data.csv")
data_raw_faked.drop("comment", axis=1, inplace = True)
print("Shape of data_raw_faked:", data_raw_faked.shape)

# APPEND FAKED data to original data
data_raw_original_and_faked = data_raw_original.append(data_raw_faked)
print("Shape of data_raw_original and faked:", data_raw_original_and_faked.shape)

####################################################################

# load Temp train data data ( unsure if this data shall be used as train data)
data_raw_temp = pd.read_csv("dataset_2020-02-19/eband_TEMP_train_data.csv")
print("Shape of data_raw_temp:", data_raw_temp.shape)


# APPEND TEMP data to original and faked data
data_raw_train = data_raw_original_and_faked.append(data_raw_temp)

print("Shape of data_raw_train", data_raw_train.shape)
print("Number of training examples (m)",data_raw_train.shape[0]/96)
```

## 2. Prepare and save EBAND DATASET

### 2.1 Augment features

```python

def augment_features(df):
    
    # IMPORTANT. DO NOT MODIFY THIS FUNCTION WITHOUT CHANGE IN ALL SCRIPTS !!!
    
    df ["rx_levels_metric"] = df ["ActualReceived_Level_x"] + df ["ActualReceived_Level_y"]
    
    df ["ModChanges_metric"] = (df ["ModChanges_x"] + df ["ModChanges_y"]) * 2
    
    df ["errors_metric"] = ((df ["SES_x"] + df ["SES_y"]) * 100) + ((df ["ES_x"] + df ["ES_y"]) * 10 )
    
    df ["UAS_metric"] = df ["UAS_x"] + df ["UAS_y"]
    
    df ["fading_vs_rx_metric"] = (-100 * df ["rx_levels_metric"]) - df ["fading_metric"]
    
    
    return df

# Call function above
eband_dataset_cnn_4 = augment_features(data_raw_train)



eband_dataset_cnn_4.shape
```

### 2.2 Fix labels

```python

def fix_labels(df):
    
    # The 15 minute Y label
    df["Y_15min"] = df["Y label"] # name in loades csv files is "Y label"
    
    df["Y_15min_text"] = "N/A"
    
    # change prediction classes from value (numbers) to text
    df.loc[df["Y label"] == 0, "Y_15min_text"]  = "0. No event"
    df.loc[df["Y label"] == 1, "Y_15min_text"]  = "1. Out of service."
    df.loc[df["Y label"] == 2, "Y_15min_text"]  = "2. HW restart."
    df.loc[df["Y label"] == 3, "Y_15min_text"]  = "3. Stuck on low modulation."
    df.loc[df["Y label"] == 4, "Y_15min_text"]  = "4. Unstable Ant. support / Wind."
    df.loc[df["Y label"] == 5, "Y_15min_text"]  = "5. Intermittent LOS problem."
    df.loc[df["Y label"] == 6, "Y_15min_text"]  = "6. LOS or Alignment problem."
    df.loc[df["Y label"] == 7, "Y_15min_text"]  = "7. Bad LOS or Alignment problem."
    df.loc[df["Y label"] == 8, "Y_15min_text"]  = "8. Rain."
    df.loc[df["Y label"] == 9, "Y_15min_text"]  = "9. Stuck on mid modulation."
    df.loc[df["Y label"] == 10, "Y_15min_text"] = "10. Wet snow."
    df.loc[df["Y label"] == 11, "Y_15min_text"] = "11. Equipment fault."
    df.loc[df["Y label"] == 12, "Y_15min_text"] = "12. Equipment fault far end."
    df.loc[df["Y label"] == 13, "Y_15min_text"] = "13. No or limited data."
    
    
    # Drop old labels
    df.drop(["Y label", "Y 2"] , axis=1, inplace=True)
    
    # Drop other columns not used
    df.drop([ "Time", "NeId", "NeType", "FailureDescription",
                                  "FarEnd_NeAlias","pred_group",
                                  "QAM1024_Strong", "QAM1024", "QAM1024_Light",
                                  "QAM2048_Strong", "QAM2048", "QAM2048_Light","QAM4096_Strong",
                                  "QAM4096", "QAM4096_Light"], axis=1, inplace=True )

    return df

eband_dataset_cnn_4 = fix_labels(eband_dataset_cnn_4)
```

### 2.2 Save dataset so it can be used by models

```python
# Save to csv, this file is used in "Visualization notebook"
eband_dataset_cnn_4.to_csv("dataset_2020-02-19/sharp_dataset_for_ai_model/eband_dataset_cnn_4.csv", sep=',', header=True, float_format='%.2f', index=False)

print("Shape of eband_dataset_cnn_4:", eband_dataset_cnn_4.shape)
```

## 3. Data visualization

## 3.1 Class distrubution 15min data label (to predict each 15 min period)

```python

def class_distr(input_df):
    df = input_df.groupby(["Y_15min_text"]).agg( { 'NeAlias': 'count',
                "sum_errors":"mean", "UAS_x": "mean", "fading_metric":"mean",
                'rx_levels_metric': "mean" , "Y_15min": "mean"} )

    df["Number of train examples"] =  df["NeAlias"]
    df["For validation/ test (80/20)"] =  df["Number of train examples"]*0.2
    df["errors_average"] = df["sum_errors"].round(1)
    df["UAS_x_average"] = df["UAS_x"].round(1)
    df["fading_metric_average"] = df["fading_metric"].round(1)
    df["rx_levels_metric_average"] = df["rx_levels_metric"].round(1)
    
    df.drop(["NeAlias", "UAS_x", "sum_errors", "fading_metric", 'rx_levels_metric'], axis=1, inplace=True)
    df.sort_values(by="Y_15min", inplace=True)
    df.reset_index(inplace=True)

    df.set_index("Y_15min_text", inplace=True)
    return df
```

```python
# Call function
view_class_distr_15min_cnn = class_distr(eband_dataset_cnn_4)
view_class_distr_15min_cnn
```

```python
def heatmap_15min_class(df):
    df = df.groupby(["Y_15min_text"]).agg({ 
                "sum_errors":"mean",
                "ActualReceived_Level_x": "mean", "ModChanges_x":"mean", "UAS_x":"mean",
                "ActualReceived_Level_y": "mean", "ModChanges_y":"mean", "UAS_y": "mean",
                "hw_fault": "mean", "fading_metric": "mean",
                'rx_levels_metric':"mean" , 'ModChanges_metric': "mean",
                   'errors_metric':"mean", "UAS_metric" : "mean"
                    })

    
    df.reset_index(inplace=True)

    df.set_index("Y_15min_text", inplace=True)
    
    # Normalize by column:
    df = (df - df.mean()) / df.std()
    
    return df

heatmap_class = heatmap_15min_class(eband_dataset_cnn_4)
```

```python
heat_map = sb.heatmap(heatmap_class , annot=True)
plt.show()
```
